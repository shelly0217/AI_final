# 切分資料集

K-fold交叉驗證

```python
from sklearn.model_selection import KFold #載入KFold套件
```

```python
# 執行K-fold交叉驗證
test_bias1 = []
test_variance1 = []
train_errors1 = []
test_errors1 = []
kf = KFold(n_splits=10, shuffle=True, random_state=42)
 
# 拆分數據集為訓練集和測試集
for train_index1, test_index1 in kf.split(x_features):
    x_train1, y_train1 = x_features.iloc[train_index1], y_label[train_index1]
    x_test1, y_test1 = x_features.iloc[test_index1], y_label[test_index1]
    
# 訓練模型
    model1.fit(x_train1, y_train1)
    
    y_train_pred1 = model1.predict(x_train1) # 預測訓練集資料
    y_test_pred1 = model1.predict(x_test1) # 預測測試集資料

# 計算偏差Bias和變異性Variance
    bias1 = np.mean((y_test_pred1 - np.mean(y_test1)) **2) / len(y_test_pred1)
    variance1 = np.var(y_test_pred1) / len(y_test_pred1)
    
    test_bias1.append(bias1)
    test_variance1.append(variance1)

# 計算訓練和測試的誤差
    train_error1 = np.sum((y_train_pred1 - y_train1) ** 2) / len(y_train_pred1)
    test_error1 = np.sum((y_test_pred1 - y_test1) ** 2) / len(y_test_pred1)

    train_errors1.append(train_error1)
    test_errors1.append(test_error1)

#輸出bias and variance
print("Bias:",test_bias1)
print("Variance:",test_variance1)
#輸出平均訓練和測試誤差
print("平均訓練誤差:", sum(train_errors1) / len(train_errors1))
print("平均測試誤差:", sum(test_errors1) / len(test_errors1))
```

偏差（Bias）：偏差表示模型在不同訓練集上的預測結果與實際值之間的差異。偏差越低，說明模型對訓練數據擬合得越好，能夠更準確地預測訓練數據的表現。在給出的結果中，平均偏差介於1.27到2.22之間，這表示模型在不同訓練集上的預測結果與實際值之間的平均差異。

變異（Variance）：變異表示模型在不同訓練集上的預測結果的穩定性。變異越低，說明模型對不同訓練數據的波動性更小，能夠更一致地預測新數據的表現。在給出的結果中，平均變異介於1.27到2.22之間，這表示模型在不同訓練集上的預測結果的平均波動性。

平均訓練誤差：平均訓練誤差表示模型在訓練集上的預測誤差的平均值。在給出的結果中，平均訓練誤差為107.06。

平均測試誤差：平均測試誤差表示模型在測試集上的預測誤差的平均值。在給出的結果中，平均測試誤差為109.76。

可以看到，模型的偏差和變異都是一定的，平均訓練誤差和平均測試誤差的值也是相近的。這表明模型可能存在一定的欠擬合，即無法很好地擬合訓練數據，同時也無法很好地泛化到新的測試數據。要改善模型的性能，可能需要考慮調整模型參數、增加更多特徵或採用其他更複雜的模型。

# 找出最適模型

使用梯度下降法進行最佳化

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

#定義梯度下降函數
def gradient_descent(X, y, learning_rate, n_iterations):
    n_samples, n_features = X.shape
    #初始化模型参數
    params = np.zeros(n_features)
    #迭代更新参數
    for _ in range(n_iterations):
        #計算預測值
        y_pred = np.dot(X, params)
        #計算梯度
        gradient = np.dot(X.T, (y_pred - y)) / n_samples
        #更新参數
        params -= learning_rate * gradient
    return params

#對特徵進行標準化
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train1)
x_test_scaled = scaler.transform(x_test1)

#添加偏置项（截距）
x_train_scaled = np.c_[np.ones(len(x_train_scaled)), x_train_scaled]
x_test_scaled = np.c_[np.ones(len(x_test_scaled)), x_test_scaled]

#設置超参數
learning_rate = 0.01
n_iterations = 1000

#使用梯度下降法训练模型
params = gradient_descent(x_train_scaled, y_train1, learning_rate, n_iterations)

#使用訓練好的参數進行預測
y_train_pred = np.dot(x_train_scaled, params)
y_test_pred = np.dot(x_test_scaled, params)

#繪制訓練數據和預測结果的散點圖
plt.scatter(x_train_scaled[:, 1], y_train1, color='blue', label='Training Data')
plt.scatter(x_test_scaled[:, 1], y_test1, color='red', label='Test Data')
plt.plot(x_train_scaled[:, 1], y_train_pred, color='green', label='Predictions')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

繪製的圖像看起來很複雜，可能是因為數據集的特徵較多或者模型的複雜度較高。在這種情況下，可能需要進一步分析和處理數據，或者選擇更簡單的模型來進行回歸分析。

```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.svm import SVR

#創建 SVR 模型
model = SVR()

#初始化儲存結果的列表
test_bias = []
test_variance = []
train_errors = []
test_errors = []

#使用 K-fold 進行交叉驗證
kf = KFold(n_splits=10, shuffle=True, random_state=42)

for train_index, test_index in kf.split(x_features):
    x_train, y_train = x_features.iloc[train_index], y_label[train_index]
    x_test, y_test = x_features.iloc[test_index], y_label[test_index]
    
    #訓練模型
    model.fit(x_train, y_train)
    
    #預測訓練集和測試集
    y_train_pred = model.predict(x_train)
    y_test_pred = model.predict(x_test)
    
    #計算偏差 Bias 和變異性 Variance
    bias = np.mean((y_test_pred - np.mean(y_test)) ** 2) / len(y_test_pred)
    variance = np.var(y_test_pred) / len(y_test_pred)
    
    #儲存偏差和變異性
    test_bias.append(bias)
    test_variance.append(variance)
    
    #計算訓練集和測試集的誤差
    train_error = np.sum((y_train_pred - y_train) ** 2) / len(y_train_pred)
    test_error = np.sum((y_test_pred - y_test) ** 2) / len(y_test_pred)
    
    #儲存訓練集和測試集的誤差
    train_errors.append(train_error)
    test_errors.append(test_error)

#輸出偏差和變異性
print("偏差（Bias）:", test_bias)
print("變異性（Variance）:", test_variance)

#輸出平均訓練誤差和平均測試誤差
print("平均訓練誤差:", np.mean(train_errors))
print("平均測試誤差:", np.mean(test_errors))
```

偏差（Bias）：偏差衡量了模型的預測值與實際值之間的平均差異程度，它反映了模型對於訓練和測試數據的適應能力。偏差越小，表示模型對於數據的擬合能力越好。在這個例子中，偏差的值介於0.15到0.28之間，整體上可以認為模型對於數據的擬合能力還是相對不錯的。
